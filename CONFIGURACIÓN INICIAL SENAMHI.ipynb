```python
# 1. Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# 'display' es útil en Jupyter, pero 'print' es más universal.
# Si estás en Jupyter, puedes añadir: from IPython.display import display

# 2. Configuración inicial
plt.style.use('seaborn-v0_8-darkgrid')
warnings.filterwarnings('ignore')

# Necesario en Jupyter para que los gráficos se muestren en el notebook
%matplotlib inline

print("Bibliotecas cargadas correctamente.")

# 3. Definir la función de visualización (MODIFICADA)
# --- (ESTA SECCIÓN NO HA CAMBIADO) ---
# Esta función ahora solo grafica la precipitación mensual.

def visualizacion_estacion(df, codigo_estacion, col_pp):
    """
    Genera y muestra las visualizaciones (Series de Tiempo, Histograma, Boxplot)
    para los datos de precipitación ACUMULADA MENSUAL de una estación.
    Añade anotaciones para Mediana, Q1 y Q3.
    """
    
    print(f"\n--- Generando Gráficos para: {codigo_estacion} (Datos Mensuales) ---")
    
    # --- Calcular estadísticas para las anotaciones ---
    try:
        # Usamos .quantile() para obtener todos los valores
        q1 = df[col_pp].quantile(0.25)
        median = df[col_pp].quantile(0.50) # O .median()
        q3 = df[col_pp].quantile(0.75)
        print(f"Estadísticas (Pp mensual): Q1={q1:.2f}, Mediana={median:.2f}, Q3={q3:.2f}")
    except Exception as e:
        print(f"Error al calcular estadísticas: {e}")
        # Si no podemos calcularlas, continuamos sin anotaciones
        q1, median, q3 = None, None, None

    # --- 1. Gráfico de Series de Tiempo (Precipitación Mensual) ---
    try:
        print(f"[Gráfico 1: Serie de Tiempo de Pp Acumulada Mensual para {codigo_estacion}]")
        fig, ax = plt.subplots(figsize=(15, 5)) # Un solo gráfico
        
        # Pp
        df[col_pp].plot(ax=ax, title=f'Precipitación Acumulada Mensual ({col_pp})', color='green', legend=True)
        ax.set_ylabel('Pp Acumulada (mm)')
        
        # --- LÍNEAS PARA LA MEDIANA ---
        if median is not None:
            # Añadir línea de mediana
            ax.axhline(y=median, color='orange', linestyle='--', label=f'Mediana: {median:.2f}')
            
            # Añadir texto de mediana (al inicio de la serie)
            ax.text(df.index[0], median, f' {median:.2f}', color='orange', 
                    va='bottom', ha='left', backgroundcolor='white', fontsize=9)
            
            ax.legend()
        # --- FIN DE LÍNEAS MEDIANA ---
            
        plt.suptitle(f'Serie de Tiempo Mensual para la Estación: {codigo_estacion}', fontsize=16, y=1.02)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error al generar serie de tiempo: {e}")

    # --- 2. Histograma (Precipitación Mensual) ---
    try:
        print(f"\n[Gráfico 2: Histograma de Distribución para {codigo_estacion}]")
        fig, ax = plt.subplots(figsize=(8, 5)) # Un solo gráfico
        
        # Pp (Mostramos todos los meses, incluidos los que tienen 0)
        sns.histplot(df[col_pp], kde=True, ax=ax, color='green')
        ax.set_title(f'Distribución de {col_pp} (Acumulado Mensual)')

        # --- LÍNEAS PARA LA MEDIANA ---
        if median is not None:
            # Añadir línea vertical de mediana
            ax.axvline(x=median, color='orange', linestyle='--', label=f'Mediana: {median:.2f}')
            
            # Añadir texto de mediana (en la parte superior del gráfico)
            ax.text(median, ax.get_ylim()[1] * 0.95, f' {median:.2f}', color='orange', 
                    ha='center', va='top', fontsize=9, backgroundcolor='white')
            
            ax.legend()
        # --- FIN DE LÍNEAS MEDIANA ---

        plt.suptitle(f'Histograma Mensual para la Estación: {codigo_estacion}', fontsize=16, y=1.02)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error al generar histograma: {e}")

    # --- 3. Boxplot (Precipitación Mensual) ---
    try:
        print(f"\n[Gráfico 3: Diagrama de Caja (Boxplot) para {codigo_estacion}]")
        fig, ax = plt.subplots(figsize=(8, 5)) # Un solo gráfico

        # Pp
        sns.boxplot(y=df[col_pp], ax=ax, color='green')
        ax.set_title(f'Boxplot de {col_pp} (Acumulado Mensual)')

        # --- LÍNEAS PARA Q1, MEDIANA Y Q3 ---
        if all([q1, median, q3]): # Si se calcularon todas las estadísticas
            # La posición x=0.05 es ligeramente a la derecha del centro (0)
            
            # Texto para Q1
            ax.text(0.05, q1, f'Q1: {q1:.2f}', color='black', va='center', ha='left', 
                    backgroundcolor='white', fontsize=9)
            
            # Texto para Mediana (Q2)
            ax.text(0.05, median, f'Mediana (Q2): {median:.2f}', color='orange', va='center', ha='left', 
                    backgroundcolor='white', fontsize=9, weight='bold')
            
            # Texto para Q3
            ax.text(0.05, q3, f'Q3: {q3:.2f}', color='black', va='center', ha='left', 
                    backgroundcolor='white', fontsize=9)
        # --- FIN DE LÍNEAS QUARTILES ---

        plt.suptitle(f'Boxplot Mensual para la Estación: {codigo_estacion}', fontsize=16, y=1.02)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error al generar boxplot: {e}")


# --- 4. Lógica Principal (Adaptada a CSV y Datos Mensuales) ---
# --- (ESTA SECCIÓN HA SIDO MODIFICADA) ---

print("Iniciando el procesamiento del archivo CSV...")

# Define el nombre de tu archivo CSV
ruta_csv = r'C:\Users\Acer\Downloads\TP2\Trabajo_Final\datos_completos_Datos_tp2.csv'

try:
    # Carga el archivo CSV completo en un DataFrame
    df_total = pd.read_csv(ruta_csv)
    
    # Fundamental: Convertir la columna 'Fecha' a tipo datetime
    df_total['Fecha'] = pd.to_datetime(df_total['Fecha'])
    
    print(f"Archivo '{ruta_csv}' cargado exitosamente.")
    
    # Obtiene la lista de estaciones únicas de la columna 'Nombre'
    nombres_estaciones = df_total['Nombre'].unique()
    
    print(f"Estaciones encontradas: {', '.join(nombres_estaciones)}")

    # Itera sobre cada nombre de estación encontrado
    for estacion in nombres_estaciones:
        
        # Muestra un título para la estación que se está procesando
        print("\n" + "="*70)
        print(f"Análisis para la Estación: {estacion}")
        print("="*70)
        
        # Filtra el DataFrame grande para obtener solo los datos de esta estación
        df_estacion_diaria = df_total[df_total['Nombre'] == estacion].copy()
        
        # Establece la 'Fecha' como índice (importante para las series de tiempo)
        df_estacion_diaria.set_index('Fecha', inplace=True)
        
        # --- ¡NUEVO FILTRO DE FECHA! ---
        # Filtramos los datos para quedarnos solo con el rango 1981-2014
        print(f"Filtrando datos entre 1981-01-01 y 2014-12-31...")
        df_estacion_diaria = df_estacion_diaria.loc['1981-01-01':'2014-12-31'] # <-- ¡NUEVO FILTRO DE FECHA!
        
        # --- ¡NUEVO PASO: MENSUALIZAR DATOS! ---
        # Resampleamos los datos diarios a mensuales ('M' = Month End)
        # y sumamos (.sum()) la precipitación de cada mes.
        
        # Seleccionamos solo la columna de precipitación
        columna_pp = 'Pp_completo'
        
        # Resampleamos y convertimos de Serie a DataFrame para mantener la estructura
        df_plot_mensual = df_estacion_diaria[columna_pp].resample('M').sum().to_frame()

        # Comprobación por si el filtro de fechas dejó el dataframe vacío
        if df_plot_mensual.empty:
            print(f"No hay datos para la estación {estacion} en el rango 1981-2014.")
            continue # Salta a la siguiente estación

        print("\n--- Primeras 5 filas de datos MENSUALES ACUMULADOS (1981-2014) ---")
        # Usamos print() para compatibilidad (en lugar de display)
        print(df_plot_mensual.head())
        
        # Muestra la información descriptiva
        print("\n--- Resumen Estadístico (Datos Mensuales Acumulados 1981-2014) ---")
        print(df_plot_mensual.describe())
        
        # Genera y muestra los gráficos para esta estación
        # (Llamada a la función simplificada)
        visualizacion_estacion(df_plot_mensual, estacion, columna_pp)

    print("\n" + "¡Proceso completado para todas las estaciones!")

except FileNotFoundError:
    print(f"ERROR: No se pudo encontrar el archivo '{ruta_csv}'.")
except Exception as e:
    print(f"Ha ocurrido un error inesperado: {e}")
```

```python
# Configuración inicial
plt.style.use('seaborn-v0_8-darkgrid')
warnings.filterwarnings('ignore')

def plot_percentiles(df, estacion, col_pp, ax):
    print(f"Graficando: {estacion}")
    
    df_p = df.groupby(df.index.year)[col_pp].quantile([0.95, 0.99]).unstack()
    df_p.columns = ['P95', 'P99']
    
    # Graficar directo en el eje 'ax'
    df_p['P95'].plot(ax=ax, label='P95', marker='.', linestyle='-', color='#4c72b0')
    df_p['P99'].plot(ax=ax, label='P99', marker='.', linestyle='--', color='#55a868')
    
    # Ajustes visuales
    ax.set_title(f'Percentiles de Precipitación Extrema\n{estacion}', fontsize=10)
    ax.set_xlabel('Año', fontsize=9)
    ax.set_ylabel('mm', fontsize=9)
    ax.legend(fontsize=8)

ruta_csv = r'C:\Users\Acer\Downloads\TP2\Trabajo_Final\datos_completos_Datos_tp2.csv'
estaciones = ['TINGOMARIA', 'PUCALLPA', 'HUANCANE', 'JAUJA']

df_total = pd.read_csv(ruta_csv)
df_total['Fecha'] = pd.to_datetime(df_total['Fecha'])

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for est, ax in zip(estaciones, axes.flatten()):
    # Filtrar datos por estación
    df_est = df_total[df_total['Nombre'] == est].copy()
    df_est.set_index('Fecha', inplace=True)  

    df_est = df_est.loc['1980':'2014']
 
    if not df_est.empty:
        plot_percentiles(df_est, est, "Pp_completo", ax)

plt.tight_layout()
plt.show() 
```
